```python
import cv2 
```

### Reading The Input

```python
image = cv2.imread("pexels-photo-842711.jpeg")
cv2.imshow("Squirrel", image)
cv2.waitKey(0)
```

###  converting the original image to grey image

```python
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
cv2.imshow("Nature", gray_image)
cv2.waitKey(0)
```

### invert the new grayscale image

```python
inverted_image = 255 - gray_image
cv2.imshow("Inverted", inverted_image)
cv2.waitKey()
```

### blur the image by using the Gaussian Function in OpenCV

```python
blurred = cv2.GaussianBlur(inverted_image, (21, 21), 0)
cv2.imshow("Blurred Image", blurred)
cv2.waitKey()
```

###  final step is to invert the blurred image

```python
inverted_blurred = 255 - blurred
pencil_sketch = cv2.divide(gray_image, inverted_blurred, scale=256.0)
cv2.imshow("Sketch", pencil_sketch)
cv2.waitKey(0)
```

#### Hence I have completely converted the RGB file into pencil sketch format
