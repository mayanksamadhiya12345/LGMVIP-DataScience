### there are so many null values but some columns are hidden , so let's unhide them firstly

```python
nan_values = df.isna()
nan_columns = nan_values.any()

col_with_nan = df.columns[nan_columns].tolist()
col_with_nan
```

### so many nan columns here so explore and process them one by one 

```python
df.rename(columns={'iyear':'Year','imonth':'Month','iday':'Day','gname':'Group','country_txt':'Country','provstate':'State',
                   'region_txt':'Region','longitude':'Longitude','summary':'Summary','attacktype1_txt':'Attacktype',
                   'targtype1_txt':'Targettype','nkill':'kill','weaptype1_txt':'Weapon',
                   'nwound':'Wound'},inplace=True)
```

```python
pd.set_option("display.max_columns",None)
print(df.dtypes)
```

#### Not all columns are necessory for us some are unnecessory so take only those columns that are necessory for analysis 

```python
df = df[['Year','Month','Day','Country','State','Region','city','latitude','Longitude',
         'Attacktype','kill','Wound','target1','Summary','Group','Targettype','Weapon','motive']]
```

### HeatMap For Visualizing The Null Values

### Conclusion Of the heatmap

as we can observe that most null values are present in    the 'motive' and 'summary' column which is also not much informative so from our dataset we can drom them
 
apart from this , 'latitude' and 'longitude' column is also not useful for the analysis part so we will remove them also

### Filling 'unknown' where there is null values in state and city

```python
df['State'].fillna("unknown" , inplace=True)
df['city'].fillna("unknown" , inplace=True)
```

### Filling '0' where null values

```python
df['Wound'] = df['Wound'].fillna(0);
df['kill'] = df['kill'].fillna(0);
df['casualities'] = df['kill'] + df['Wound']
```

```python
df.head()
```

### Drop all columns with null values

```python
df.dropna(axis=1,inplace=True)
```

## DATA VISUALIZATION ON TERRORISM DATASET 

#### countrywise attacks

```python
plt.rcParams["figure.figsize"] = 8 , 5
sns.barplot(x=df["Country"].value_counts()[:20].index,y=df["Country"].value_counts()[:20].values)
plt.ylabel("Number Of Total Attacks" , fontsize=17)
plt.xticks(rotation=90)
plt.xlabel("Attack Region" , fontsize=17)
plt.title("Country-Wise Attacks Bar-Plot" , fontsize=25 , fontweight="bold")
```

#### monthwise attacks 

```python
plt.rcParams["figure.figsize"] = 8 , 5
sns.barplot(x=df["Month"].value_counts()[:20].index,y=df["Month"].value_counts()[:20].values)
plt.ylabel("Number Of Total Attacks" , fontsize=17)
plt.xticks(rotation=90)
plt.xlabel("Attack Month" , fontsize=17)
plt.title("Month-Wise Attacks Bar-Plot" , fontsize=25 , fontweight="bold")
```

#### attacks in different regions

```python
pd.crosstab(df.Year,df.Region).plot(kind='area',stacked=False)
plt.title("Terrorist Activities By Regions In Each Year",fontsize=24)
plt.ylabel("Number Of Attacks",fontsize=18)
plt.xlabel("Year",fontsize=18)
plt.show()
```

#### Number Of Attacks Over Different Cities 

```python
df['city'].value_counts().to_frame().sort_values('city',axis=0,ascending=False).head(10).plot(kind='bar',color='r')
plt.xticks(rotation=50)
plt.xlabel("city",fontsize=15)
plt.ylabel("Number Of Attacks",fontsize=15)
plt.title("Top 10 Most Effected Cities",fontsize=20,fontweight="bold")
plt.show()
```

#### Attacks By Different Groups

```python
plt.rcParams["figure.figsize"] = 8 , 5
sns.barplot(x=df["Group"].value_counts()[:20].index,y=df["Group"].value_counts()[:20].values)
plt.ylabel("Number Of Total Attacks" , fontsize=17)
plt.xticks(rotation=90)
plt.xlabel("Attack Group" , fontsize=17)
plt.title("Attack By Group Bar-Plot" , fontsize=25 , fontweight="bold")
```

#### Number Of Kills by Each Terrorist Organisation in Each Country 

```python
group = df[['Group','Country','kill']]
group = group.groupby(['Group','Country'],axis=0).sum().sort_values('kill',ascending=False).drop('Unknown').reset_index().head(10)
group
```

#### Number Of People That Killed By the Different Types OF attack

```python
df[['Attacktype','kill']].groupby(['Attacktype'],axis=0).sum().plot(kind='bar',figsize=(10,7),color='r')
plt.xticks(rotation=50)
plt.title("Number Of Killed",fontsize=18)
plt.ylabel("Number Of People",fontsize=18)
plt.xlabel=("Attac Type")
plt.show()
```

#### Total Number Of People That Killed By the Attack in Each Country
